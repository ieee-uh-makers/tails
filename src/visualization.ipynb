```python
%matplotlib inline

import matplotlib
import numpy as np
import csv
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = [10., 5.]
matplotlib.rcParams["font.size"] = "22"
matplotlib.rcParams["lines.linewidth"] = "3"
matplotlib.rcParams["axes.grid"] = True
```

```python
# Plot Results

pwms = {"ail": "pwm_7.csv", 
        "ele": "pwm_8.csv",
        "thr": "pwm_11.csv",
        "rud": "pwm_12.csv"}

mintime = None

for pwm in pwms:
    dat = []
    f = open(pwms[pwm], 'r')
    c = csv.DictReader(f)
    for row in c:
        if mintime is None:
            mintime = float(row['time'])
        elif float(row['time']) < mintime:
            mintime = float(row['time'])
        break
        


for pwm in pwms:
    dat = []
    
    f = open(pwms[pwm], 'r')
    c = csv.DictReader(f)
    for row in c:
        dat.append([float(row['time']), float(row['duty'])])

    xy = np.array(dat)
            
    x = xy[:,0] - mintime
    y = xy[:,1]
    
    miny = np.min(y)
        
    plt.scatter(x, y, linewidths=8)
    plt.title("Duty Cycles: %s" % pwm)
    plt.xlabel("Time (s)")
    plt.ylabel("Duty Cycle")
    f = open("fsm.csv", 'r')
    
    c = csv.DictReader(f)
    for row in c:
        if mintime is None:
            mintime = float(row['time'])
        elif float(row['time']) < mintime:
            mintime = float(row['time'])
            
        
        plt.annotate(row['state'], (float(row['time']) - mintime, miny), arrowprops=dict(facecolor='black', shrink=0.05))
    plt.show()
    



```
